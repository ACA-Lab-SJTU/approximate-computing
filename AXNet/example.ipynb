```python
from build import Model
import tensorflow as tf
from error import error_func
import numpy as np

bench_mark = 'inversek2j'
total_iteration = 50000
total_LR =  1e-4
error_bound = 0.1
sess = tf.InteractiveSession()
err_f = error_func(bench_mark)


acc_net_list = [2,5,2]
c_net = [2,3,7]
model = Model(sess, bench_mark, acc_net_list, c_net, error_bound, err_f, total_LR)
print(model.result_dir)
model.train(total_iteration)
```

```python
re = []
ab = []
tr = []
ci = []
x, y, clf_y = model.create_clf_data_label(train=False, final=True)
if model.keep_tail:
    clf_if, accu = model.sess.run([model.clf_if_flag, model.accuracy],
                                 feed_dict={model.inputs: x, model.label: y})
    clf_y = np.argmax(clf_y, 1)

    tr.append(np.sum(clf_y) / clf_y.size)
    ci.append(np.sum(clf_if) / clf_if.size)

    re.append(np.sum([1 if clf_y[i] == clf_if[i] else 0 for i in range(clf_if.size)]) / clf_if.size)
    ab.append(np.mean(np.multiply(clf_if, np.mean(accu, 1))))
    print(np.multiply(clf_if, np.mean(accu, 1)).size)
    print("Now we have tail!!" + '\n')
    print("True invocation: \t\t\t\t" + str(np.mean(tr)) + '\n')
    print("CLF invocation: \t\t\t\t" + str(np.mean(ci)) + '\n')
    print("Final CLF accuracy: \t\t\t" + str(np.mean(re)) + '\n')
    print("Error in CLF's prediction: \t\t" + str(np.mean(ab)) + '\n')
```
